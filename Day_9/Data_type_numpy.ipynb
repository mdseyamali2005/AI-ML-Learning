```python
import numpy as np
```

**int32, int 64 , float32 , float 64, complex64, complex128, s(byte string) , u(unicode string), object**

```python
arr = np.array([2,3,4,6,3,2])
print(arr, arr.dtype)
```

**downcast**

```python
downcast = arr.astype(np.int16)
print(downcast, downcast.dtype)
```

```python
arr = np.array([2,3,4,6,3.1,2])
print(arr, arr.dtype)
```

```python
arr2 = np.array([2,3,4, "seyam", {3,34,6}])
print(arr2, arr2.dtype)
```

**complex number**

```python
arr3 = np.array([3+6j])
arr4 = np.array([4+2j])
print(arr3, arr4 , arr3.dtype, arr4.dtype)
print(arr3 + arr4)
print(arr3 - arr4)
arr5 = arr3 + arr4
print(arr5, arr5.dtype)
```
