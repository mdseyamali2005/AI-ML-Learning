***Numpy vs python list (time and memory)***

```python
import numpy as np
import time
```

***Time***

```python
size = 10_000_000
py_list = list(range(size))
start = time.time()
square_of_all_py_list_value = [i**2 for i in py_list]
end = time.time()
print(f"need time to square of py list= {end - start} second")
numpy_arr = np.array(py_list)
start = time.time()
square_of_all_numpy_arr_value = numpy_arr**2
end = time.time()
print(f"need time to square of numpy arr= {end - start} second")
```

***Memory***

```python
import sys #sys represent system module
print(f"the memory size of py list {sys.getsizeof(py_list) * len(py_list)} bytes") #py list just reference store kore
print(f"the memory size of numpy arr {numpy_arr.nbytes} bytes")
a = 10
b = [10]
c = [1,2,3]
print(sys.getsizeof(a))
print(sys.getsizeof(b))
print(sys.getsizeof(c))
```
