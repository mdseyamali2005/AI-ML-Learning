**IN python list slicing make actual copy**

**in numpy slicing make just view**

```python
list = [3,5,7,8,9,3,5,90]
print(len(list))
slicing = list[2:6]
print(slicing)
slicing[2]= 500
print(slicing)
print(list)
```

**in numpy create just view**

```python
import numpy as np
```

```python
arr= np.array([3,5,7,8,9,3,5,90])
slicing = arr[2:6]
print(slicing)
slicing[2]= 500
print(slicing)
print(arr)
```
