```python
import numpy as np
```

```python
arr1 = np.zeros((2,3))
print(arr1, type(arr1), arr1.dtype, arr1.shape)
```

```python
arr2 = np.zeros((2,3), dtype = "int64")
print(arr2, type(arr2), arr2.dtype, arr2.shape)
```

```python
arr3 = np.ones((3,4))
print(arr3, arr3.shape)
```

```python
arr4 = np.full((2,3),49)
print(arr4, arr4.dtype, arr4.shape)
```

```python
arr5 = np.eye((3), dtype = "int32")
print(arr5, arr5.shape)
```

```python
arr6 = np.arange(1,11)
print(arr6, arr6.shape)
```

```python
arr7 = np.arange(1,11, 2)#increment by 2
print(arr7, arr7.shape)
```

```python
arr8 = np.linspace(1,100, 5)#1 thake 100 projonto but 5 vag e vag kore 5 ta value dia array banabe
print(arr8, arr8.shape)
```
