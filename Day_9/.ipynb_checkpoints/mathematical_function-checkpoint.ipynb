```python
import numpy as np
```

```python
arr = np.array([2,4,5,2,4,67,6,7,8,23,45,67,23,98,34])
```

**Mathematical Function**

```python
print("sum or all value= ", np.sum(arr))
print("multiplication of all value(if by default int32)= ", arr.prod())
arr2 = arr.astype(np.int64)
print("multiplication of all value(int64)= ", arr2.prod())
print("min value from array= ", np.min(arr))
print("max value from array= ", np.max(arr))
print("min value index from array= ", np.argmin(arr))
print("max value index from array= ", np.argmax(arr))
print("mean from array= ", np.mean(arr))
print("median from array= ", np.median(arr))
print("standerd deviation from array= ", np.std(arr))
print("variance(variance = (standard deviation)²) from array= ", np.var(arr))
```

**POWER**

```python
print("square all value from array= ", np.square(arr))
print("sqrt all value from array= ", np.sqrt(arr))
print("pow all value from array= ", np.pow(arr, 3)) #arr.pow(3)
```

*#arr.pow(3) dile amara python er functin use kortechi , but np.--- dile amara numpy er function use kortechi*

**Log and Exponential**

```python
print("log all value from array= ", np.log(arr))
print("log all value from array= ", np.log10(arr))
print("log all value from array= ", np.log2(arr))
print("exp all value from array= ", np.exp(arr))
```

**Rounding**

```python
arr3 = np.array([3.43,24.235, 234, 4.56, 5.5, 8.9])
print("rounding the value= ", np.round(arr3)) #.5 and up the next value and below .5 is the current value
print("floor the value= ", np.floor(arr3))#all time lower value
print("ceil the value= ", np.ceil(arr3))#all time higher value
print("trunc value= ", np.trunc(arr3)) #after the decimal value cut
```
