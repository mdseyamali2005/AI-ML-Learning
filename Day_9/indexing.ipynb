```python
import numpy as np
```

```python
arr = np.array([8,9,6,4,2,6,3,9])
print(arr[3])
```

```python
arr = np.array([[8,9,6],[6,3,7]])
print(arr, arr[0][2])
```

**Fancy indexing**

```python
arr = np.array([8,9,6,4,2,6,3,9])
ind = [0,2,4]
print(arr[ind])
```

```python
arr = np.array([[8,9,6],[6,3,7]])
ind = [0,0,1],[0,2,1]
print(arr, arr[ind])
```

**condition**

```python
arr = np.array([8,9,6,4,2,6,3,9])
print(arr[arr>4])
```

```python
arr = np.array([8,9,6,4,2,6,3,9])
print(arr[arr %2 ==0])
print(arr[arr %2 !=0])
```
