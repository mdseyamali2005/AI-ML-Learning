```python
oscar_movie_name = [
    "Avatar", 
    "Titanic", 
    "The Lord of the Rings: The Return of the King", 
    "Oppenheimer", 
    "Forrest Gump"
]

oscar_year = [1997,2003, 2009, 2023, 2025]

oscar_earning = [2923, 2264, 1147, 977, 678]
#-----------------------------------------------

non_oscar_movie_name = [
    "Avengers: Endgame", 
    "Star Wars: Episode VII - The Force Awakens", 
    "Jurassic World", 
    "The Avengers", 
    "The Super Mario Bros. Movie"
]

non_oscar_year = [2009, 2010, 2015, 2020, 2023]

non_oscar_earning = [2799, 2071, 1671, 1520, 1361]
```

*The oscar vs non oscar movie vusialization*

```python
plt.plot(oscar_year,oscar_earning, label = "oscar movie")
plt.plot(non_oscar_year,non_oscar_earning, label = "non oscar movie")
plt.title("oscar vs non oscar movie")
plt.xlabel("year")
plt.ylabel("earning in milion $")
plt.legend()
```

**change the line format strings**

```python
plt.plot(oscar_year, oscar_earning, color="red", marker="<", linestyle="-.", linewidth=5, label ="Oscar Movies") 
plt.plot(non_oscar_year, non_oscar_earning, label="Non-Oscar Movies") 
plt.title("Oscar Movies vs Non-Oscar Movies Revenue") 
plt.xlabel("Year") 
plt.ylabel("Revenue (in $M)") 
plt.legend()
```
