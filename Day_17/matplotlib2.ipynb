```python
import matplotlib.pyplot as plt
import numpy as np
```

**oscar and non oscar movie comparision**

```python
oscar_movie_name = [
    "Avatar", 
    "Titanic", 
    "The Lord of the Rings: The Return of the King", 
    "Oppenheimer", 
    "Forrest Gump"
]

years = [1997,2003, 2009, 2023, 2025]

oscar_earning = [2923, 2264, 1147, 977, 678]
#----------------------------------------------------------
non_oscar_movie_name = [
    "Avengers: Endgame", 
    "Star Wars: Episode VII - The Force Awakens", 
    "Jurassic World", 
    "The Avengers", 
    "The Super Mario Bros. Movie"
]

non_oscar_earning = [2799, 2071, 1671, 1520, 1361]
```

```python
x = np.arange(len(years))
weidth = 0.4
plt.bar(x-weidth/2, oscar_earning,weidth, label= "oscar movie")
plt.bar(x+weidth/2, non_oscar_earning,weidth, label= "non oscar movie")
plt.title("oscer vs non oscar movie reveniow")
plt.xlabel("years")
plt.ylabel("reveniow in milion $")
plt.legend()
plt.xticks(x,years)
```
