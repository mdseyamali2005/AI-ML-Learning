```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
# Task 
days = ["Mon", "Tue", "Wed", "Thu", "Fri"] 
cities = ["New York", "London", "Delhi", "Tokyo"] 
temperatures = [ [22, 23, 21, 24, 25], # New York 
                  [18, 19, 17, 20, 21], # London 
                  [30, 32, 31, 33, 34], # Delhi 
                  [25, 26, 24, 27, 28] # Tokyo 
                 ]
```

**General form manully**

```python
fig, ax= plt.subplots(2,2)
ax[0][0].plot(days, temperatures[0])
ax[0][0].set_title(cities[0])

ax[0][1].plot(days, temperatures[1])
ax[0][1].set_title(cities[1])

ax[1][0].plot(days, temperatures[2])
ax[1][0].set_title(cities[2])

ax[1][1].plot(days, temperatures[3])
ax[1][1].set_title(cities[3])

plt.tight_layout()
```

**Using nested loop**

```python
fig, axes=plt.subplots(2, 2)
city_no = 0 
for i in range(2): # rows (i) - 0,1 
    for j in range(2): # cols(j) -0,1 
        axes[i][j].plot(days, temperatures [city_no], marker='o') #marker o oi point gula te mark kore diche
        axes[i][j].set_title(cities [city_no]) 
        axes[i][j].grid(True) 
        city_no = city_no + 1
plt.tight_layout()
```

**2D array use na kore 1D array te kori, more easyer**

```python
fig, axes= plt.subplots (2, 2) 
axes = axes.flatten() #flatten() 2D array-কে 1D array বানিয়ে দেয়।
for i,ax in enumerate (axes): #enumerate() → Loop করার সময় একই সাথে index (i) এবং element (ax) দেয়।
    ax.plot(days, temperatures [i], marker='o') 
    ax.set_title(cities [i]) 
    ax.grid(True)
fig.suptitle("Temperature in Cities over the week") 
fig.supylabel("Temperature") 
fig.supxlabel("Days") 
fig.tight_layout()
```
