```python
import numpy as np
import pandas as pd
```

```python
iris = pd.read_csv("IRIS.csv")
iris
```

```python
print(iris[:10])
print("------")
print(iris.dtypes)
print("------")
print(iris.shape)
print("------")
print(iris.describe().loc[["mean","std","min","max"]])
```

```python
print(iris[iris["petal_length"]>4.5])
print("------")
print(iris[iris["species"] == "Iris-virginica"])
print("------")
print(iris[(iris["petal_length"]>4.5) & (iris["species"]=="Iris-virginica")])
```

```python
print(iris.groupby("species")["sepal_length"].mean())
print("------")
print(iris.groupby("species")["petal_width"].max())
print("------")
print(iris.groupby("species")["sepal_width"].std())
iris["petal_ratio"] = iris["petal_length"]/iris["petal_width"]
iris.groupby("species")["petal_ratio"].mean()
```
