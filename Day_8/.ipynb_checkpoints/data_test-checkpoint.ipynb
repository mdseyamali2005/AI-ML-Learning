```python
import json
```

```python
def data_in(Name):
    with open(Name, "r") as f:
        data = json.load(f)
        return data
```

```python
data = data_in("data.json")
print(data)
```

```python
def clean_data(data):
    text_to_number = {"one": 1, "two": 2, "three": 3, "four": 4, "five": 5 }
    final_user = []
    unick_user = set()
    for value in data:
        row_rating = str(value["rating"]).strip().lower()
        if row_rating in text_to_number:
            row_rating = text_to_number[row_rating]
        else:
            try:
                row_rating = float(row_rating)
            except:
                row_rating = None   # invalid case যেমন "excellent"
        value["rating"] = str(row_rating)
        #Handle the missing value of age
        row_age = value.get("age")
        value["age"] = row_age
        if(value["name"] in unick_user):
            continue
        unick_user.add(value["name"])
        final_user.append(value)
    return final_user
```

```python
clean_data(data)
```
