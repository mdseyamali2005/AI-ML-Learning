```python
import json
```

```python
def data_in(Name):
    with open(Name, "r") as f:
        data = json.load(f)
        return data
```

```python
data = data_in("data.json")
print(data)
```

```python
def clean_data(data):
    text_to_number = {"one": 1, "two": 2, "three": 3, "four": 4, "five": 5 }
    for value in data:
        row_rating = value["rating"].strip().lower()
        if row_rating in text_to_number:
            row_rating = text_to_number[row_rating]
        print(row_rating)
        value["rating"] = row_rating
    for user in data:
        print(user)
```

```python
clean_data(data)
```
