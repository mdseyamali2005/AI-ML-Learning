```python
import requests
```

```python
url = "https://quickchart.io/chart"

# Chart এর configuration (Chart.js format)
chart_config = {
    "type": "bar",
    "data": {
        "labels": ["January", "February", "March"],
        "datasets": [{
            "label": "Sales",
            "data": [10, 20, 30]
        }]
    }
}

# Parameters পাঠাও
params = {
    "chart": str(chart_config).replace("'", '"'),  # JSON string
    "width": 500,
    "height": 300,
    "backgroundColor": "white"
}

response = requests.get(url, params=params)
#                             ↑         ↑
#                       keyword arg   তোমার variable dictionary
print(response.status_code)  # 200 হবে

# Image save করো
if response.status_code == 200:
    # with দিলে কাজ শেষে File বন্ধ করতে হয় na,  automatically বন্ধ হয়ে যায়।
    with open("chart.png", "wb") as f:
        f.write(response.content) #chat.png er modhe content dia diche
    print("chart.png saved!")
```
