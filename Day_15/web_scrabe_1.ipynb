```python
from bs4 import BeautifulSoup
with open("scrabe_file/first_scrabe.html", "r") as f:
    html_content = f.read()
soup = BeautifulSoup(html_content, "lxml")
```

```python
all_h3 = soup.find_all("h3")
```

```python
country_name = []
for i in all_h3:
    country_name.append(i.get_text(strip= True))
country_name
```
