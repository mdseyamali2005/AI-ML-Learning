```python
import pandas as pd
```

```python
df = pd.read_csv("globalAirQuality.csv")

print("--- Rows where AQI > 100 ---")
print(df[df["aqi"] > 100])

print("\n--- AQI > 100 AND Temperature > 30 ---")
print(df[(df["aqi"] > 100) & (df["temperature"] > 30)])

print("\n--- AQI > 100, Temp > 30, and Country is 'US' ---")
print(df[((df["aqi"] > 100) & (df["temperature"] > 30)) & (df["country"] == "US")])

print("\n--- Country and City for high AQI areas ---")
print(df[df["aqi"] > 100][["country", "city"]])

data = df[(df["aqi"] > 100) & (df["temperature"] > 30)][["country", "city"]]

print("\n--- First row of filtered data (position-based) ---")
print(data.iloc[0]) # Always the first row of the NEW subset

print("\n--- Row with Index Label 6 (label-based) ---")

print(data.loc[6]) 

print("\n--- The resulting 'data' DataFrame ---")
print(data)
```
