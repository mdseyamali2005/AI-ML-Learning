```python
import pandas as pd
```

**Using Dictionary**

```python
info = {
    "Name" : ["Seyam", "Bob", "doo"],
    "Age" : [34,23,12],
    "CGPA" : [3.4,3.5,3.7]
}
dataframe = pd.DataFrame(info)
print(dataframe , type(dataframe))
# print(dataframe[1])
print(dataframe.index)
print(dataframe.columns)
```

**List of list**

```python
data = pd.DataFrame([["seyam",32],["bob",31],["kamal",12]],columns = ["Name", "age"])
print(data)
```
