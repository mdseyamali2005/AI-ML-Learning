```python
import pandas as pd
```

```python
df = pd.read_csv("globalAirQuality.csv")
print("--- First 5 Rows ---")
print(df.head())

print("\n--- Last 5 Rows ---")
print(df.tail())

print("\n--- Summary Statistics ---")
print(df.describe()) #if we dont use print function then code return only list 
```

**use loc for level and iloc for index**

```python
# Selection by label (loc)
print("--- Row 0 (loc) ---")
print(df.loc[0])

print("\n--- Rows 0 to 3 (loc is inclusive) ---")
print(df.loc[0:3]) # in loc end index are inclusive

# Selection by position (iloc)
print("\n--- Row at index 2 (iloc) ---")
print(df.iloc[2])

print("\n--- Rows 0 to 3 (iloc is exclusive) ---")
print(df.iloc[0:3]) # in iloc end index are exclusive

# Column selection
print("\n--- Specific Column (o3) ---")
print(df["o3"]) # if we want to print specific colum

print("\n--- Multiple Columns ---")
print(df[["country","aqi"]])

# Mixed selection (Rows and Columns)
print("\n--- Rows 0-2 for specific columns ---")
print(df.loc[0:2, ["country","aqi"]])

print("\n--- Specific Cell (loc) ---")
print(df.loc[2,"o3"])

print("\n--- Specific range using index values ---")
print(df.iloc[0:3, [0,1,2]]) # when we use indexing then we need to give columns index value

print("\n--- Row 2 with specific column indices ---")
print(df.iloc[2, [2,3,4,5]])

# Scalar access (faster for single values)
print("\n--- Fast Access (at) ---")
print(df.at[1,"country"])

print("\n--- Fast Access (iat) ---")
print(df.iat[2,4])

# Deep Copy vs View
print("\n--- Copying and Manipulating ---")
dataCopy = df["city"].copy() # without copy it only view , and with copy we can edit and manupulate any thing
dataCopy[1]= "mumby"
print(dataCopy[0:4])

# Printing the whole DataFrame
print("\n--- Full DataFrame ---")
print(df) # here also , but without removing the comment
```
