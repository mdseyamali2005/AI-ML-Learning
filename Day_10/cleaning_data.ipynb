```python
import pandas as pd
```

```python
df = pd.read_csv("raw_data.csv")
print(df.isnull(),"\n") #/df.isna()
print(df.isna().sum(),"\n")
print(df.dropna(),"\n") #drop row wise
print(df.dropna(axis = 1),"\n") # colume wise drop
print(df.fillna(0),"\n") #fill NAN value with given value
copy_data = df.copy()
age_mean = df["age"].mean()
copy_data["age"] = df["age"].fillna(age_mean) #fill nan value with the mean for age
copy_data = copy_data.ffill() #filling forword , up to dow , and fill nan value with previous value
print(copy_data,"\n")
print(df.bfill(),"\n") #filling backward , dow to up , and fill nan value with previous value
print(copy_data.isna().sum())
```

```python
print(df.duplicated()) #for row wise
print()
print(df["country"].duplicated()) #for colum wise  
recover_data = df.copy() 
print()
print(df.drop_duplicates(inplace = True)) #if we use inplace = True then change the df
print()
print(df["country"].drop_duplicates())
print()
print(df)
```
